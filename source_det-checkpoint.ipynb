```python
from astropy.stats import sigma_clipped_stats
from photutils.datasets import make_100gaussians_image
from photutils.detection import find_peaks
from astropy.stats import sigma_clipped_stats
from astropy.io import fits
from astropy.table import Table
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from astropy.visualization import simple_norm
from astropy.visualization.mpl_normalize import ImageNormalize
from photutils.aperture import CircularAperture
```

```python
hdu = fits.open(r"/home/riya/Desktop/gal_fit/nircam6/crop1/f115w1.fits")  
data = hdu[0].data 
mean, median, std = sigma_clipped_stats(data, sigma=18.0)
threshold = median + (29.0 * std)
tbl = find_peaks(data, threshold, box_size=15)
tbl['peak_value'].info.format = '%.8g' # for consistent table output
print(tbl)
```

```python
df = tbl.to_pandas()

# Save the DataFrame to an Excel file
df.to_csv("/home/riya/Desktop/gal_fit/nircam6/crop1/test.csv", index=False)
```

```python
positions = np.transpose((tbl['x_peak'], tbl['y_peak']))
apertures = CircularAperture(positions, r=5.0)
norm = simple_norm(data, 'sqrt', percent=99.9)

plt.figure(figsize=(15, 15))  # Increase the figure size (width, height)
plt.imshow(data, cmap='Greys_r', origin='lower', norm=norm, interpolation='nearest')
apertures.plot(color='#0547f9', lw=1.5)
plt.xlim(0, data.shape[1] - 1)
plt.ylim(0, data.shape[0] - 1)
plt.show()
```
